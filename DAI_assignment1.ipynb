About the Dataset

*   Distance_from_Home: Distance from home to school (Near, Moderate, Far).
*   Gender: Student's gender (Male, Female).
*   Exam_Score: Student's final exam score.
*  Hours_Studied: Weekly study hours spent by the student.
*  Attendance: Percentage of attended classes.
*  Parental_Involvement: Degree of parental involvement in the student's education (Low, Medium, High).
*  Access_to_Resources: Availability of educational resources (Low, Medium, High).
*  Extracurricular_Activities: Participation in extracurricular activities (Yes, No).
*   Sleep_Hours: Average nightly sleep duration (in hours).
*   Previous_Scores: Scores from prior exams.
*   Motivation_Level: Student's motivation level (Low, Medium, High).
*  Tutoring_Sessions: Number of monthly tutoring sessions attended.
*   Internet_Access: Availability of internet access (Yes, No).
*  Family_Income: Family's income level (Low, Medium, High).
*   Teacher_Quality: Quality rating of the student's teachers (Low, Medium, High).

```python
df.dropna(inplace = True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Outlier Detection and Treatment
# Using IQR method for numerical features
print("\nHandling Outliers...")
num_cols = df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    print(f"{col}: {outlier_count} outliers detected.")
    df[col] = np.where(df[col] > upper_bound, upper_bound,
                       np.where(df[col] < lower_bound, lower_bound, df[col]))
```

```python
# Standardizing Categorical Values
df['Parental_Involvement'] = df['Parental_Involvement'].str.strip().str.capitalize()
df['Motivation_Level'] = df['Motivation_Level'].str.strip().str.capitalize()
df['Internet_Access'] = df['Internet_Access'].str.strip().str.capitalize()
```

```python
# Histograms and Boxplots
for col in num_cols:
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')

    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

Bivarate Analysis

```python
# Scatter plots

for col in df.select_dtypes(include = "number").columns.drop(['Exam_Score']):
  plt.figure(figsize=(6, 4))
  plt.scatter(df[col], df['Exam_Score'], alpha=0.5, color='blue', edgecolors='black')
  plt.title(f'Scatter plot of {col} vs Exam_Score')
  plt.xlabel(col)
  plt.ylabel('Exam_Score')
  plt.grid(True)
  plt.show()
```

```python
# Encode Categorical Variables for Correlation Analysis
le = LabelEncoder()
categorical_cols = [
    'Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities',
    'Motivation_Level', 'Internet_Access', 'Family_Income', 'Teacher_Quality',
    'School_Type', 'Peer_Influence', 'Learning_Disabilities',
    'Parental_Education_Level', 'Distance_from_Home', 'Gender']

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter Plots for Numerical Variables
for col in df.select_dtypes(include="number").columns.drop(['Exam_Score']):
    plt.figure(figsize=(6, 4))
    plt.scatter(df[col], df['Exam_Score'], alpha=0.5, color='blue', edgecolors='black')
    plt.title(f'Scatter plot of {col} vs Exam_Score')
    plt.xlabel(col)
    plt.ylabel('Exam_Score')
    plt.grid(True)
    plt.show()
```

```python
# Bar plots and violin plots for categorical vs numerical
plt.figure(figsize=(10, 5))
sns.boxplot(x='Parental_Involvement', y='Exam_Score', data=df)
plt.title('Parental Involvement vs Exam Score')
plt.show()

plt.figure(figsize=(10, 5))
sns.violinplot(x='Motivation_Level', y='Exam_Score', data=df)
plt.title('Motivation Level vs Exam Score')
plt.show()
```
